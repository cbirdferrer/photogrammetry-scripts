## Set Up
if you want to run this yourself you'll need a specific file structure

* Main Folder
    * measurements _keep this folder empty, the code will save files here_
    * csvs _this folder is where each students outputs from the GUI should be saved_
        * Student1
            * Whale_1.csv
            * Whale_2.csv
    * rankings _this folder is where you should save each student's ranking notes, make sure the name matches the name of the folder in csvs_
        * Student1_rankings.csv
    * plots _keep this folder empty, the script will save all the plots it outputs here_
    * outputs _keep this folder empty, the script will save the output summary tables here_
    * secret.csv _if you have a key of which whale corresponds to which ID, save it here_
    
    
we designed this for what worked best for our lab, however, you are free to set up your own naming system, just be sure to edit the script accordingly

## Part 1: Collate all the GUI outputs as a loop

```python
#first import the necessary modules and set up the directory
import pandas as pd
import os,sys
import numpy as np
import math

wd = r'~file path to main folder here~' 

csvfolder = os.path.join(wd, "csvs")
rankfolder = os.path.join(wd,'rankings')
folders = os.listdir(csvfolder) #make list of all folders in the csv folder
print(folders)
for folder in folders: 
    f = os.path.join(csvfolder,folder) 
    ff = os.path.isdir(f) #make sure the folder is a folder
    #print(ff)
    if ff == False:
        continue
    elif ff == True:
        #print('YAY')
        GUIfold = f
        print(GUIfold)
        files = os.listdir(GUIfold)
        measurements= []
        nonPercMeas = []

        def anydup(l): #we'll use this function later to check for duplicates
            seen = set()
            for x in l:
                if x in seen: return True
                seen.add(x)
            return False
        #here we loop through all the csvs to extract all the possible measurement names
        #if the file starts with a number and ends with a .csv, use it
        for f in (f for f in files if f.endswith('.csv')):
            ff = os.path.join(GUIfold, f)
            #print(ff)
            df0 = pd.read_csv(ff,sep = ',', usecols = [0],skip_blank_lines = True)
            #display(df0)
            idx = df0.loc[df0['Image ID'] == 'Object Name'].index
            #print(idx[0])
            df = pd.read_csv(ff, sep = ",",header = (idx[0]+1),skip_blank_lines = True)
            #display(df)
            
            colList = df.columns.tolist()
            if 'Object Name' in colList:
                col1 = 'Object Name'
                widths = df.columns.values.tolist()
            else:
                for i in colList:
                    values = df[i].tolist()
                    #print(values)
                    if 'Object Name' in values:
                        col1 = i
                idxON = df.loc[df[col1] == 'Object Name'].index[0]
                new_header = df.iloc[idxON]
                df = df[1:]
                df.columns = new_header
                widths = df.columns.values.tolist()
            #print(col1)
            
            
            l = df[col1].tolist() #make a list of all the measurement names in this column
            #print(l)
            
            l = [x for x in l if pd.isna(x) == False] #elimate all empty cells
            if 'Object Name' in l: l.remove('Object Name') #get rid of Object Name since it's not a measurement
            measurements = measurements + l #add the measurment names to the master list
            nonPercMeas = nonPercMeas + l #copy of the master list that does not include widths
            
            df = df.set_index('Object Name')

            if anydup(l) == True:
                print("please check file {0} for duplicate Object Names and remove duplicates".format(f))
                sys.exit("run script again when you get rid of the duplicate measurement type")
            elif anydup(l) == False:
                #print(f)
                for i in l: #loop through list of measurement types
                    for w in (w for w in widths if w[0].isdigit()): #loop through the widths
                        x = df.loc[i,w] #extract the value of the cell of the width of the measurement
                        if pd.isna(x) == False: #if that cell isn't empty i.e width was measured
                            ww = i + "_" + w #combine the name of the measurement type with the width
                            measurements += [ww] #add this combined name to the master list
                            #print(x)

        #now we're going to set up a dictionary to fill in with all the measurements
        #that we will eventually turn into a dataframe where the keys are the columns
        rawM = measurements
        rawMM = set(rawM)
        measurements += ['Image', 'Animal_ID','Altitude','Focal Length'] #add the Image name and ID to the dictionary
        mDict = dict.fromkeys(measurements) #make a dictionary with just the keys, no values yet
        keys = list(mDict.keys())

        df_all = pd.DataFrame(columns = keys) #set up the empty dataframe using the keys from the dict


        for f in (f for f in files if f.endswith('.csv')):
            file = os.path.join(GUIfold, f)
           # print(f)
            df0 = pd.read_csv(file,sep = ',',header = 0, usecols = [0])
            idx = df0.loc[df0['Image ID'] == 'Object Name'].index
            #display(df0)
            
            df = pd.read_csv(file, sep = ',', header = None, nrows = idx[0])
            #display(df)
            wID = df.at[0,1] #extract the whale ID cell value
            aID = f.split(".")[0] 
            aID = "W" + aID[1:]
            mDict['Animal_ID'] = f.split(".")[0] #add to the dictionary
            image = df[df[0] == 'Image Path'].loc[:,[1]].values[0]
            #print(image)
            mDict['Image'] = image
            focl = float(df[df[0] == 'Focal Length'].loc[:,[1]].values[0])
            mDict['Focal Length'] = focl
            alt = float(df[df[0] == 'Altitude'].loc[:,[1]].values[0])
            mDict['Altitude'] = alt
            #print(mDict['Altitude'])
            pixd = float(df[df[0] == 'Pixel Dimension'].loc[:,[1]].values[0])
            
            
            dfGUI = pd.read_csv(file, sep = ',',header = idx[0]+1)
            #print(file)
            #display(dfGUI)
            cols = dfGUI.columns.tolist()
            if 'Object Name' in cols:
                dfGUI = dfGUI.set_index('Object Name')
            else:
                for i in cols:
                    values = dfGUI[i].tolist()
                    if 'Object Name' in values:
                        idxCol = i
                dfGUI = dfGUI.set_index(idxCol)
            #print(dfGUI.index)
            
            secretcsv = os.path.join(wd,'secret.csv')
            df_sec = pd.read_csv(secretcsv,sep = ',',usecols = ['Animal_ID','Focal Length_Corr','Altitude_Corr'])
            #display(df_sec)
            #print(aID)
            foc_act = float(df_sec[df_sec.Animal_ID == aID].loc[:,'Focal Length_Corr'].values[0])
            alt_act = float(df_sec[df_sec.Animal_ID == aID].loc[:,'Altitude_Corr'].values[0])
            
            #print()
            
            for key in keys: #loop through the keys aka future column headers
                if key in nonPercMeas: #if that key is in the list of measurement types (not widths)
                    #print(key)
                    if key in dfGUI.index: #if that key is in the csv of this particular image
                        xx = float(dfGUI.loc[key, 'Length']) #extract the measurement value using its location
                        pixc = (xx/pixd)*(focl/alt)
                        x = ((alt_act/foc_act)*pixd)*pixc
                    else: #if this key is not in the csv
                        x = np.nan #make its value nan
                    #print(x)
                    mDict[key] = x #add the value to it's respective key
                elif "%" in key and key.split("_")[0] in dfGUI.index: #if the key is a width measurement
                    #print (key)
                    row = key.split("_")[0] #split the name of the measurement
                    col = key.split("_")[1] #to get the row and column indices
                    yy = float(dfGUI.loc[row, col]) #to extract the measurement value
                    
                    pixc = (yy/pixd)*(focl/alt)
                    y = ((alt_act/foc_act)*pixd)*pixc
                    #print(y)
                    mDict[key] = y #add the value to its respective key
            #print(mDict)
            df_op = pd.DataFrame(data = mDict, index = [1]) #make dictionary into a dataframe
            df_all = pd.concat([df_all,df_op], sort = True) #add this dataframe to the master dataframe
            #print("CONCATENATED")
    
               
    cols = list(df_all) #make list of all column headers
    a = "WwIiTtRrAaBbCcDdEeFfGgHhJjKkLlMmNnOoPpQqSsUuVvXxYyZz" #make your own ordered alphabet
    col = sorted(cols, key=lambda word:[a.index(c) for c in word[0]]) #sort headers based on this alphabet
    df_all = df_all.loc[:,col] #sort columns based on sorted list of column header    
    
    df_all.columns = df_all.columns.str.replace(".00%", ".0%")
    
    #calculate body volume
    volm = ['20.0%', '30.0%', '40.0%', '50.0%', '60.0%', '70.0%', '80.0%']
    vlist = []
    for i in volm:
        #print(i)
        for ii in df_all.columns:
            #print(ii)
            if i in ii:
                vlist += [ii]

    ids = []
    vs = []
    for i,j in enumerate(vlist[:-1]):
        jj = vlist[i+1]
        for r, R, hh,anid in zip(df_all[j],df_all[jj], df_all['Total Length'],df_all['Animal_ID']):
            h = float(hh)*.10
            r = float(r)/float(2)
            R = float(R)/float(2)
            v1 = (1/3)*math.pi*h*((r**2)+(r*R)+(R**2))
            
            ids += [anid]
            vs += [v1]

    d = {'Animal_ID':ids, 'Body_Vol':vs}
    df = pd.DataFrame(data = d)
    df1 = df.groupby('Animal_ID').sum().reset_index()
    
    #import ranking csv
    measurer = os.path.split(GUIfold)[1]
    rankcsv = os.path.join(rankfolder, "{0}_rankings.csv".format(measurer))
    df_rank = pd.read_csv(rankcsv, sep = ",", usecols = ['Animal_ID','Rank','Experience Level 1-4'])
    
    df_all1 = pd.merge(df_all,df1, on = ['Animal_ID'])
    df_all1 = pd.merge(df_all1,df_rank, on = ['Animal_ID'])
    df_all1['Measurer'] = measurer
    
    savefold = os.path.join(wd,"measurements")
    outcsv = os.path.join(savefold,"{0}_measurements.csv".format(measurer))
    df_all1.to_csv(outcsv,sep = ',')

print("DONE")
```

## Part 2: Combine each student's measurement csv into master sheet
to do this I first get the column names from just one of the student combined csvs and make an emtpy dataframe with those column names

then I loop through each student combined csv and concatenate (add vertically) that dataframe to the empty one I made

then I remove some unnecessary columns

### Combining Data Frames

```python
allfold = os.path.join(wd,'measurements')
files = os.listdir(allfold)

for i in files:
    file = os.path.join(allfold,i)
    df1 = pd.read_csv(file, sep = ",")
    cols = df1.columns
    break

df_master = pd.DataFrame(columns = cols)

for f in files:
    print(f)
    file = os.path.join(allfold,f)
    dfx = pd.read_csv(file,sep = ",")
    #print(dfx)
    df_master = pd.concat([df_master,dfx],sort = True)
    
df_master = df_master.drop(['Image', 'Unnamed: 0'], axis = 1)

df_master['idx'] = [int(x.split("-")[1]) for x in df_master['Animal_ID']]
df_master = df_master.sort_values(by = ['idx'])

#add in the secret data
sfile = os.path.join(wd,"secret.csv")
df_s = pd.read_csv(sfile,sep = ",")

df_master = pd.merge(df_master,df_s,how = "left", on = ['Animal_ID'])
```

### Check that the altitude and focal length entered by the student were correctly entered

```python
focalList = []
for fc,fs in zip(df_master['Focal Length_Corr'],df_master['Focal Length']):
    fc = float(fc)
    if fs != fc:
        f = "wrong"
        focalList +=[f]
    elif fs == fc:
        f = "yay"
        focalList += [f]

altlist = []
for ac,als in zip(df_master['Altitude_Corr'],df_master['Altitude']):
    if ac != als:
        a = "wrong"
        altlist += [a]
    elif ac == als:
        a = "yay"
        altlist += [a]
        
df_master['AltCheck'] = altlist
df_master['FLcheck'] = focalList
display(df_master[:5])
```

```python
#export final master dataframe as csv
outloc = os.path.join(wd,"outputs","mastersheet_allmerged.csv")
df_master.to_csv(outloc,sep=",")
```

### look at number of wrong altitudes or focal lengths per image

```python
import seaborn as sns
df_wrong = df_master.loc[df_master['AltCheck'] == 'wrong']
df_w = df_wrong.groupby('idx')['AltCheck'].count().reset_index().rename(columns = {"AltCheck": 'count'})
df_w['category'] = 'Altitude'

df_wr = df_master.loc[df_master['FLcheck'] == 'wrong']
dfww = df_wr.groupby('idx')['FLcheck'].count().reset_index().rename(columns = {'FLcheck': 'count'})
dfww['category'] = 'Focal Length'

df_bad = pd.concat([df_w,dfww])
df_bad
p = sns.barplot(x = 'idx', hue = 'category',y = 'count', data = df_bad)
plt.suptitle("Number of incorrect entries per image")
fig = os.path.join(figfold, "incorrect_entries_perimg.jpg")
plt.savefig(fig,dpi = 500)
plt.show()
```

### calculate the summary statistics

```python
df_master['Rank'] = [int(x) for x in df_master['Rank']]

df_r = df_master.groupby(['idx'])['Rank'].mean().reset_index()


df_mTL = df_master.groupby(['idx'])['Total Length'].agg(['min','max','mean','std','var']).reset_index()
df_mTL['cv'] = (df_mTL['std']/df_mTL['mean'])*100
df_mTL = pd.merge(df_mTL,df_r, on = ['idx'])
df_mF = df_master.groupby(['idx'])['Fluke'].agg(['min','max','mean','std','var']).reset_index()
df_mF['cv'] = (df_mF['std']/df_mF['mean'])*100
df_mF = pd.merge(df_mF,df_r, on = ['idx'])
df_mV = df_master.groupby(['idx'])['Body_Vol'].agg(['min','max','mean','std','var']).reset_index()
df_mV['cv'] = (df_mV['std']/df_mV['mean'])*100
df_mV = pd.merge(df_mV,df_r, on = ['idx'])
print("total length")
display(df_mTL)
print("")
print("fluke")
display(df_mF)
print("")
print("volume")
display(df_mV)

#make csvs
od = os.path.join(wd,"outputs")
tl = os.path.join(od,"tl_byimg.csv")
df_mTL.to_csv(tl, sep = ',')
fl = os.path.join(od,"fluke_byimg.csv")
df_mF.to_csv(fl,sep = ",")
vol = os.path.join(od,"vol_byimg.csv")
df_mV.to_csv(vol,sep = ",")
```
